```python
import re
import json
!pip install PyPDF2
from PyPDF2 import PdfReader

# === LOAD PDF ===
pdf_path = "Midlands Rules of Evidence.pdf"
reader = PdfReader(pdf_path)
full_text = "\n".join([page.extract_text() for page in reader.pages])

# === CLEAN & NORMALIZE ===
full_text = re.sub(r"\n+", "\n", full_text)
full_text = re.sub(r"\s+", " ", full_text)
full_text = re.sub(r"Rule\s+(\d{3})\s*\.", r"Rule \1.", full_text)

# === SPLIT INTO RULES ===
raw_rules = re.split(r"(Rule \d{3}\. )", full_text)[1:]
combined_rules = [raw_rules[i] + raw_rules[i + 1] for i in range(0, len(raw_rules), 2)]

# === PARSE EACH RULE ===
rules = []
for raw in combined_rules:
    match = re.match(r"Rule (\d{3})\. ([^.]+)\. (.+)", raw)
    if not match:
        continue

    rule_id, title, body = match.groups()
    body = body.strip()

    subparts = re.findall(r"\(([a-z])\)\s+(.*?)(?=(\s\([a-z]\)\s+)|$)", body)
    if subparts:
        text = {label: part.strip() for label, part, _ in subparts}
    else:
        text = body

    rules.append({
        "id": rule_id,
        "title": f"Rule {rule_id}. {title.strip()}",
        "text": text
    })

# === SAVE AS JSON ===
output_file = "midlands_rules.json"
with open(output_file, "w") as f:
    json.dump(rules, f, indent=2)

print(f"✅ Extracted {len(rules)} rules.")
```
